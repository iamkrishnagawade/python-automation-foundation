# Abstract Class

Am abstract class is a class that cannot be represent directly. <br>

It use to define :

- Common structure
- Common methods
- But without full implementation 

### Python Module for Abstract Classes

``` from abc import ABC, abstractmethod ```

```ABC``` -> Abstract Class <bR>
```@abstractmethod``` -> Marks a method as a abstract 

### Creating an Abstract Class

````python
from abc import ABC, abstractmethod

class Animal(ABC): # Animal abstract
    
    @abstractmethod
    def sound(self): # abstract - must be implemented by child class
        pass
````

### Child Class must implement abstract methods

````python
class Dog(Animal):
    def sound(self):
        print("Bark")

dog = Dog()
dog.sound()
````

If child does NOT implement:

````python
class Cat(Animal):
    pass

cat = Cat()
````

### Abstract Class with Concrete + Abstract Methods

````python
class Vehical:
    def start(self): # concrete
        print("Starting")

    @abstractmethod
    def wheels(self): # abstract class
        pass
````

Child Class:

````python
class Bike(Vehical):
    def wheels(self):
        return 2
````

### Why Abstract Classes?

- Enforce a common structure
- Avoid incomplete implementation
- Template for all children

### Abstract Properties

````python
class Shape():
    @property
    @abstractmethod
    def area(self):
        pass
````

````python
class Circle(Shape):
    @property
    def area(self):
        return 3.14 * 5 * 5
````
